## Research Question

How have employment size and annul median wages in selected U.S. occupations changed from 2019 to 2025, 
and do these trends significantly differ between high- and low-AI-exposure occupations?

This project focuses on whether selected occupations with higher AI exposure show different employment and 
wage trends compared with selected occupations with lower AI exposure. We are not trying to claim that AI 
directly causes unemployment or job loss.

## Background and Prior Work

Artificial intelligence is becoming more common in many types of work, especially through tools that can help
with writing, coding, debugging, explaining code, generating documentation, customer service, and data
analysis. Because of this, many students, workers, and employers are interested in how AI may change job
opportunities, wages, and the skills needed in the labor market.

A useful way to study this topic is through AI exposure. In this project, AI exposure means how much an
occupation’s tasks may be affected, assisted, or changed by AI tools. Prior work by Eloundou et al. studies
how large language models may affect the U.S. labor market by evaluating occupational tasks for LLM
exposure. Their study suggests that many workers may have at least some tasks affected by LLMs, but it
focuses on task exposure rather than claiming that entire occupations will disappear (Eloundou et al.).

This distinction is important for our project. A high AI exposure score does not mean that the job will
disappear. For example, software developers may have high AI exposure because AI tools can assist with
coding tasks, but human workers may still be needed to design systems, review code, solve complex problems,
and communicate with teams. Similarly, lower-AI-exposure jobs may still be affected indirectly by technology
or the broader economy.

To study employment and wage trends, we plan to use data from the U.S. Bureau of Labor Statistics
Occupational Employment and Wage Statistics program. BLS OEWS provides employment and wage
estimates for many occupations, which makes it useful for comparing occupation-level labor market trends
over time (“Occupational Employment and Wage Statistics Overview”).

Our project will compare selected high-AI-exposure occupations with selected low-AI-exposure occupations
from 2019 to 2025. This makes the project more specific than studying the entire labor market, while still
allowing us to compare trends before and after the rise of generative AI. Since the May 2025 OEWS data are
scheduled to be released in May 2026, we may include 2025 if the data are released and can be processed
before the final project deadline (“Schedule of Selected Releases 2026”). If not, we will use 2019 to 2024 data
and describe this limitation.

## Hypothesis


We hypothesize that high-AI-exposure occupations may show different employment and wage trends from
low-AI-exposure occupations after the rise of generative AI. However, we do not necessarily expect higher AI
exposure to always be associated with worse job outcomes. AI may change job tasks by assisting workers
rather than fully replacing them, so some high-exposure occupations may still have stable employment or
wage growth.

We also expect high-AI-exposure occupations to have higher wages on average, because many of these jobs
involve technical, analytical, or information-processing work. However, employment trends may vary
depending on industry demand, broader economic conditions, and how AI tools are adopted in different
occupations.

## Data

For this project proposal, our ideal dataset would combine AI exposure scores with occupation-level labor
market data for selected U.S. occupations. Each row would represent one occupation in one year, identified by
occupation title and SOC code.

We plan to use BLS OEWS data from 2019 to 2025 if the 2025 data are released and can be processed in time.
Since the final project is due after the scheduled release date of the May 2025 OEWS data, we may be able to
include 2025. However, if the 2025 data are not available in time or are too difficult to process, we will use
2019 to 2024 data instead and describe this as a limitation.

The selected occupations would include both high-AI-exposure and low-AI-exposure occupations. High-
exposure occupations may include jobs involving coding, data analysis, writing, or information processing.
Low-exposure occupations may include more hands-on, physical, or in-person service jobs. However, we will
try to define high and low exposure groups using an existing AI exposure score dataset instead of relying only
on our own judgment.

Possible variables would include:

| Variable | Description |
| :--- | :--- |
| Year | The year of the observation |
| Occupation name | The title of the occupation |
| SOC code | Standard Occupational Classification code used to identify occupations |
| AI exposure score | A measure of how much the occupation's tasks may be affected by AI |
| AI exposure group | High or low AI exposure group based on AI exposure scores |
| Employment size | Estimated number of workers in that occupation |
| Median annual wage | Median annual wage for that occupation |
| Mean annual wage | Mean annual wage for that occupation |
| Occupation category or industry group | Broader occupation category or industry group |

The data would likely be stored in a tabular format, such as CSV files or pandas DataFrames. Because this is a
longitudinal project, the final dataframe would use occupation-year observations. For example, “Software
Developers in 2019” would be one row, and “Software Developers in 2020” would be another row. At this
proposal stage, we do not need to have the final dataset yet, but we will need to find and verify the actual data
before the first project checkpoint.

Possible data sources may include BLS OEWS occupation data and existing AI exposure datasets from prior
research.

## Team Expectations 

Team Expectation 1: Communication

We will use our group chat as our main communication channel and use shared documents or GitHub for
project materials. Each member should check messages regularly and respond within 24 hours when possible.
If someone is busy or unable to finish a task on time, they should tell the group as early as possible.

Team Expectation 2: Respectful discussion

We will communicate respectfully, especially when giving feedback or disagreeing with an idea. We will try
to be direct but polite, explain our reasoning, and assume that criticism is meant to improve the project rather
than attack anyone personally.

Team Expectation 3: Fair contribution

Each member is expected to contribute fairly to the project. Work may be divided based on strengths and
availability, but everyone should contribute to some part of topic discussion, data work, coding, visualization,
writing, or editing.

Team Expectation 4: Task management and deadlines

We will divide tasks clearly before each checkpoint and agree on internal deadlines before the official
deadlines. We will keep track of progress through group messages, shared documents, or GitHub so that
everyone knows what still needs to be done.

Team Expectation 5: Check-ins and final review

Before each major submission, we will check the work together to make sure it is complete, clear, and
consistent. No one should make major changes right before submission without informing the group.

Team Expectation 6: Handling problems or conflicts

If someone is stuck or cannot complete their assigned work, they should ask for help early. If conflicts happen,
we will first discuss them respectfully as a group and try to find a solution. If a serious contribution problem
continues, we will document the issue and contact the professor or TA if needed.
