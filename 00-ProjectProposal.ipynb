## Research Question


How have employment size and annual median wages in selected U.S. occupations changed from 2019 to 2025, and do these trends significantly differ between high- and low-AI-exposure occupations?

This project focuses on whether selected occupations with higher AI exposure show different employment and wage trends compared with selected occupations with lower AI exposure. We are not trying to claim that AI directly causes unemployment or job loss.


## Background and Prior Work


Artificial intelligence is becoming more common in many types of work, especially through tools that can help with writing, coding, debugging, explaining code, generating documentation, customer service, and data analysis. Because of this, many students, workers, and employers are interested in how AI may change job opportunities, wages, and the skills needed in the labor market.

A useful way to study this topic is through AI exposure. In this project, AI exposure means how much an occupation's tasks may be affected, assisted, or changed by AI tools. Prior work by Eloundou et al. studies how large language models may affect the U.S. labor market by evaluating occupational tasks for LLM exposure. Their study suggests that many workers may have at least some tasks affected by LLMs, but it focuses on task exposure rather than claiming that entire occupations will disappear (Eloundou et al.).

This distinction is important for our project. A high AI exposure score does not mean that the job will disappear. For example, software developers may have high AI exposure because AI tools can assist with coding tasks, but human workers may still be needed to design systems, review code, solve complex problems, and communicate with teams. Similarly, lower-AI-exposure jobs may still be affected indirectly by technology or the broader economy.

To study employment and wage trends, we plan to use data from the U.S. Bureau of Labor Statistics Occupational Employment and Wage Statistics program. BLS OEWS provides employment and wage estimates for many occupations, which makes it useful for comparing occupation-level labor market trends over time ("Occupational Employment and Wage Statistics Overview").

Our project will compare selected high-AI-exposure occupations with selected low-AI-exposure occupations from 2019 to 2025. This makes the project more specific than studying the entire labor market, while still allowing us to compare trends before and after the rise of generative AI. Since the May 2025 OEWS data are scheduled to be released in May 2026, we may include 2025 if the data are released and can be processed before the final project deadline ("Schedule of Selected Releases 2026"). If not, we will use 2019 to 2024 data and describe this limitation.

## Hypothesis


We hypothesize that high-AI-exposure occupations may show different employment and wage trends from low-AI-exposure occupations after the rise of generative AI. However, we do not necessarily expect higher AI exposure to always be associated with worse job outcomes. AI may change job tasks by assisting workers rather than fully replacing them, so some high-exposure occupations may still have stable employment or wage growth.

We also expect high-AI-exposure occupations to have higher wages on average, because many of these jobs involve technical, analytical, or information-processing work. However, employment trends may vary depending on industry demand, broader economic conditions, and how AI tools are adopted in different occupations.


## Data


For this project proposal, our ideal dataset would combine AI exposure scores with occupation-level labor market data for selected U.S. occupations. Each row would represent one occupation in one year, identified by occupation title and SOC code.

We plan to use BLS OEWS data from 2019 to 2025 if the 2025 data are released and can be processed in time. Since the final project is due after the scheduled release date of the May 2025 OEWS data, we may be able to include 2025. However, if the 2025 data are not available in time or are too difficult to process, we will use 2019 to 2024 data instead and describe this as a limitation.

The selected occupations would include both high-AI-exposure and low-AI-exposure occupations. High-exposure occupations may include jobs involving coding, data analysis, writing, or information processing. Low-exposure occupations may include more hands-on, physical, or in-person service jobs. However, we will try to define high and low exposure groups using an existing AI exposure score dataset instead of relying only on our own judgment.

Possible variables would include:

| Variable | Description |
|---|---|
| Year | The year of the observation |
| Occupation name | The title of the occupation |
| SOC code | Standard Occupational Classification code used to identify occupations |
| AI exposure score | A measure of how much the occupation's tasks may be affected by AI |
| AI exposure group | High or low AI exposure group based on AI exposure scores |
| Employment size | Estimated number of workers in that occupation |
| Median annual wage | Median annual wage for that occupation |
| Mean annual wage | Mean annual wage for that occupation |
| Occupation category or industry group | Broader occupation category or industry group |

The data would likely be stored in a tabular format, such as CSV files or pandas DataFrames. Because this is a longitudinal project, the final dataframe would use occupation-year observations. For example, "Software Developers in 2019" would be one row, and "Software Developers in 2020" would be another row. At this proposal stage, we do not need to have the final dataset yet, but we will need to find and verify the actual data before the first project checkpoint.

Possible data sources may include BLS OEWS occupation data and existing AI exposure datasets from prior research.


## Proposed Analysis


Our analysis will mainly use exploratory data analysis, trend visualization, and simple statistical comparisons. We do not plan to use complicated machine learning models.

First, we would clean and combine BLS OEWS data across years. We would check missing values, make variable names consistent, convert wage and employment variables into usable numeric formats, and match occupations across years using SOC codes when possible. We would also merge in AI exposure scores and classify occupations into high- and low-AI-exposure groups.

Then, we would create visualizations to compare trends over time. Possible graphs include line charts of employment size from 2019 to 2025, line charts of median wage over time, and group comparisons between high- and low-AI-exposure occupations. We may also compare average wage and employment changes before and after the rise of generative AI.

For a simple before-and-after comparison, we may treat 2019 to 2022 as the pre-generative-AI period and 2023 to 2025 as the early post-generative-AI period.

Finally, we may calculate percent changes in employment and wages, compare group averages, or run simple regression models if the data are suitable. Our goal is to describe trends and associations, not to make strong causal claims.


## Ethics & Privacy

This project has relatively low privacy risk because we plan to use public occupation-level data, not individual-level data. We will not collect personal names, resumes, individual salaries, private employment records, or private company data.

The main ethical concern is interpretation. AI exposure should not be treated as the same thing as job loss risk. A high AI exposure score means that some tasks in an occupation may be affected or assisted by AI, but it does not mean the occupation will disappear. We will be careful to describe our results as trends and associations, not causal proof.

There may also be bias in the data or in the way AI exposure is measured. National-level occupation averages may hide differences by location, company size, worker experience, education level, and specific skills. Labor market trends may also be affected by factors other than AI, such as COVID-19 effects, interest rates, tech layoffs, outsourcing, industry demand, and broader economic conditions. We will discuss these limitations when communicating our results.
