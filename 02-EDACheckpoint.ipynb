# Research Question

Among U.S. occupations, do those with above-median AI exposure, as measured by the Felten et al. AIOE index, experience higher rates of median wage growth and lower rates of employment growth from 2019 to 2025 compared to occupations with below-median AI exposure?

## Background and Prior Work

Artificial intelligence is becoming more common in many types of work, especially through tools that can help with writing, coding, debugging, explaining code, generating documentation, customer service, and data analysis. Because of this, many students, workers, and employers are interested in how AI may change job opportunities, wages, and the skills needed in the labor market.

A useful way to study this topic is through AI exposure. In this project, AI exposure means how much an occupation's tasks may be affected, assisted, or changed by AI tools. Prior work by Eloundou et al. studies how large language models (LLMs) may affect the U.S. labor market by evaluating occupational tasks for their alignment with LLM capabilities. Their findings reveal that around 80% of the U.S. workforce could have at least 10% of their work tasks affected by the introduction of LLMs, while approximately 19% of workers may see at least 50% of their tasks impacted.


While Eloundou et al. provide a foundational framework for identifying which jobs are most exposed to generative AI, their study is inherently forward-looking and focuses strictly on theoretical task exposure. The authors themselves acknowledge a major limitation in their methodology: their exposure measure simply reflects an estimate of the technical capacity to make human labor more efficient. They note that "social, economic, regulatory, and other determinants imply that technical feasibility does not guarantee labor productivity or automation outcomes".

This distinction is what inspired our research question. Prior work effectively measures what could happen based on task characteristics, but it is missing an empirical look at what is actually happening in the real world over time. A high AI exposure score does not mean that a job will disappear. For example, software developers may have high AI exposure because AI tools can assist with coding tasks, but human workers may still be needed to design systems, review code, solve complex problems, and communicate with teams. Similarly, lower-AI-exposure jobs may still be affected indirectly by technology or the broader economy.

To fill this gap and test whether theoretical exposure translates into measurable outcomes, we plan to use data from the U.S. Bureau of Labor Statistics Occupational Employment and Wage Statistics (OEWS) program. BLS OEWS provides employment and wage estimates for many occupations, which makes it useful for comparing occupation-level labor market trends over time ("Occupational Employment and Wage Statistics Overview").

While Eloundou et al. focus specifically on LLMs, we will utilize the Artificial Intelligence Occupational Exposure (AIOE) dataset by Felten et al. as our primary empirical proxy. The AIOE provides a robust, standardized measure of how various AI capabilities link to specific occupational abilities, making it highly compatible with BLS data.

Our project will compare selected high-AI-exposure occupations with selected low-AI-exposure occupations from 2019 to 2025. This makes the project more specific than studying the entire labor market, while still allowing us to track actual employment size and wage trends before and after the rise of generative AI to see if the theoretical exposure translates to real-world impact. # Hypothesis


We hypothesize that occupations with above-median AI exposure, as measured by the Felten et al. AIOE index, will experience higher rates of median wage growth from 2018 to 2024 compared to occupations with below-median AI exposure. This is because many high-AI-exposure occupations involve technical, analytical, writing, coding, or information-processing tasks that often require specialized skills and may benefit from AI-assisted productivity gains.

We also hypothesize that occupations with above-median AI exposure will experience lower rates of employment growth from 2018 to 2024 compared to occupations with below-median AI exposure. This is because some tasks in high-exposure occupations may be easier for AI tools to assist, automate, or change, which could reduce the need for additional workers in some occupations.

However, we will interpret these results as trends and associations rather than causal proof. Other factors, such as COVID-19 effects, inflation, interest rates, tech layoffs, outsourcing, industry demand, and broader economic conditions, may also affect wage and employment growth during this period.

**Description of the variables most relevant to this project:**

* **`OCC_TITLE` (String):** The descriptive title of the occupation.

2. **Lack of AI Metrics:** This dataset contains strictly labor market metrics and has no native measure of AI exposure, necessitating a merge with a secondary dataset.
3. **Changes in SOC Definitions:** While not a massive issue for a single year, as we expand this data back to 2019, we may encounter occupations where the SOC codes or definitions were updated, requiring manual alignment.

```python
## YOUR CODE TO LOAD/CLEAN/TIDY/WRANGLE THE DATA GOES HERE
import pandas as pd
import numpy as np

file_path = 'data/00-raw/bls_national_2023.xlsx'
raw_bls_df = pd.read_excel(file_path)

cols_to_keep = ['OCC_CODE', 'OCC_TITLE', 'O_GROUP', 'TOT_EMP', 'A_MEDIAN', 'A_MEAN']
bls_df = raw_bls_df[cols_to_keep].copy()

print(f"Original dataset size: {bls_df.shape[0]} rows and {bls_df.shape[1]} columns.")
display(bls_df.head())
```

The original BLS dataset contains aggregate occupation categories (e.g., 'total', 'major', 'broad') alongside specific occupations ('detailed'). To make the dataset tidy and avoid double-counting employees, we filtered the data to only include 'detailed' occupations. Now, each row represents a single, unique occupation observation.

```python
bls_tidy = bls_df[bls_df['O_GROUP'] == 'detailed'].copy()
bls_tidy = bls_tidy.drop(columns=['O_GROUP'])

print(f"Tidy dataset size (only detailed occupations): {bls_tidy.shape[0]} rows.")
```

In the BLS dataset, missing values are not initially represented as NaN, but rather as specific string characters like * (data not available or suppressed for data quality) and # (wage is equal to or greater than $239,200/year). The missingness here is systematic: BLS suppresses data when sample sizes are too small to protect employer privacy, or when wages exceed their top coding threshold.

```python
print("Count of suppressed/missing values (represented by '*' or '#'):")
for col in ['TOT_EMP', 'A_MEDIAN', 'A_MEAN']:
    star_count = bls_tidy[col].astype(str).str.contains(r'\*').sum()
    hash_count = bls_tidy[col].astype(str).str.contains(r'\#').sum()
    print(f"{col} - '*' count: {star_count}, '#' count: {hash_count}")
```

We replaced all BLS suppression symbols (*, #, **) with NaN. We chose to use dropna(how='any') to remove these rows.

Since our research question requires comparing median wages and employment size with AI exposure scores, keeping rows with missing wages or employment numbers would break our analysis. Because the dataset is large enough, dropping these rows provides the cleanest set for regression and EDA.

```python
pd.set_option('future.no_silent_downcasting', True)
bls_tidy.replace(['*', '#', '**'], np.nan, inplace=True)

for col in ['TOT_EMP', 'A_MEDIAN', 'A_MEAN']:
    bls_tidy[col] = bls_tidy[col].astype(str).str.replace(',', '', regex=False)
    bls_tidy[col] = pd.to_numeric(bls_tidy[col], errors='coerce')

bls_clean = bls_tidy.dropna(how='any')

print(f"Final cleaned dataset size: {bls_clean.shape}")
```


After summarizing the data, we observed huge variance in employment size (TOT_EMP). Some specialized occupations have very few workers, while others (like 'Fast Food and Counter Workers') have millions. We do not consider these as "suspicious" or errors; they reflect the true inequality of the U.S. labor market. We will keep them but may use logarithmic scales during our EDA phase.

```python
display(bls_clean.describe())

print("\nTop 3 occupations with highest employment:")
display(bls_clean.nlargest(3, 'TOT_EMP'))
```

### Dataset #2: Artificial Intelligence Occupational Exposure (AIOE) - Language Modeling

* **Dataset Name:** AIOE - Language Modeling Scores (Felten et al., 2021)
* **Link to the dataset:** https://github.com/AIOE-Data/AIOE
* **Number of observations:** 774 occupation-level observations in the original AIOE dataset; 668 observations remain after merging with the cleaned BLS dataset.
* **Number of variables:** 3 original variables relevant to our analysis; after cleaning, we add one derived variable, ai_exposure_group.

**Description of the variables most relevant to this project:**
* **`SOC Code` (String):** The Standard Occupational Classification code. This serves as the primary key to link the AI exposure scores with the BLS labor metrics.
* **`Occupation Title` (String):** The official name of the occupation.
* **`Language Modeling AIOE` (Numeric):** A standardized z-score representing an occupation's exposure to Large Language Models (LLMs). A score of 0 represents the average exposure across all occupations; a positive score indicates above-average exposure (high potential for AI assistance/augmentation), while a negative score indicates below-average exposure.

**Descriptions of any shortcomings this dataset has with respect to the project:**
1. **Theoretical Nature:** The AIOE index measures technical exposure—the theoretical capacity for AI to perform occupational tasks—rather than actual real-world adoption or job displacement rates.
2. **Version Mismatch:** The AIOE scores are based on an older SOC taxonomy. This resulted in the loss of approximately 130 occupations during the merge process, as some newer or updated 2023 BLS job codes did not have a direct match in the AIOE dataset.
3. **Subjectivity in Mapping:** The scores are derived from mapping AI applications to human abilities via human and algorithmic labeling, which may introduce subjective bias or fail to capture the complexity of high-level human interpersonal tasks.

```python
aioe_raw = pd.read_excel(
    "data/00-raw/Language Modeling AIOE and AIIE.xlsx",
    sheet_name="LM AIOE"
)

print("Dataset shape:", aioe_raw.shape)
print("Columns:", aioe_raw.columns.tolist())

aioe_raw.head()
```

```python
aioe_clean = aioe_raw[[
    "SOC Code",
    "Occupation Title",
    "Language Modeling AIOE"
]].copy()

aioe_clean = aioe_clean.rename(columns={
    "SOC Code": "soc_code",
    "Occupation Title": "occupation_title",
    "Language Modeling AIOE": "aioe_score"
})

aioe_clean.head()
```

```python
aioe_clean["soc_code"] = aioe_clean["soc_code"].astype(str).str.strip()
aioe_clean["occupation_title"] = aioe_clean["occupation_title"].astype(str).str.strip()
aioe_clean["aioe_score"] = pd.to_numeric(aioe_clean["aioe_score"], errors="coerce")

aioe_clean.dtypes
```

```python
print("Missing values:")
display(aioe_clean.isna().sum())

print("Duplicated rows:", aioe_clean.duplicated().sum())
print("Duplicated SOC codes:", aioe_clean["soc_code"].duplicated().sum())
```

```python
aioe_clean = aioe_clean.dropna(subset=["soc_code", "occupation_title", "aioe_score"])
aioe_clean = aioe_clean.drop_duplicates()

print("Cleaned dataset shape:", aioe_clean.shape)

print("Missing values after cleaning:")
display(aioe_clean.isna().sum())

print("Duplicated rows after cleaning:", aioe_clean.duplicated().sum())
print("Duplicated SOC codes after cleaning:", aioe_clean["soc_code"].duplicated().sum())
```

```python
median_aioe = aioe_clean["aioe_score"].median()

aioe_clean["ai_exposure_group"] = np.where(
    aioe_clean["aioe_score"] > median_aioe,
    "above_median",
    "below_median"
)

print("Median AIOE score:", median_aioe)
aioe_clean["ai_exposure_group"].value_counts()
```

After checking Dataset #2, we confirmed that the AIOE dataset is tidy for our project. Each row represents one occupation, each column represents one variable, and each cell contains one value. We kept the three variables needed for our analysis: SOC code, occupation title, and Language Modeling AIOE score.

To address the Data Checkpoint feedback, we checked missing values, duplicated rows, duplicated SOC codes, and data types. We also converted the AIOE score into a numeric variable and created an above-median / below-median AI exposure group based on the median AIOE score. This cleaned version of Dataset #2 can be merged with the BLS OEWS dataset using SOC code.

```python
bls_clean.loc[:, "OCC_CODE"] = bls_clean["OCC_CODE"].astype(str).str.strip()

merged_df = pd.merge(
    bls_clean,
    aioe_clean,
    left_on="OCC_CODE",
    right_on="soc_code",
    how="inner"
)

print(f"Merge successful. Combined dataset contains {len(merged_df)} rows.")

merged_df[[
    "OCC_CODE",
    "OCC_TITLE",
    "TOT_EMP",
    "A_MEDIAN",
    "aioe_score",
    "ai_exposure_group"
]].head()
```

```python
print("Missing values in merged dataset:")
display(merged_df[[
    "OCC_CODE",
    "OCC_TITLE",
    "TOT_EMP",
    "A_MEDIAN",
    "aioe_score",
    "ai_exposure_group"
]].isna().sum())

print("Duplicated rows in merged dataset:", merged_df.duplicated().sum())
```
