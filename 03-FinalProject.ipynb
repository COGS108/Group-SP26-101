## Research Question

Among U.S. occupations, do those with above-median AI exposure, as measured by the Felten et al. AIOE index, experience higher rates of median wage growth and lower rates of employment growth from 2019 to 2025 compared to occupations with below-median AI exposure?

## Background and Prior Work

Artificial intelligence is becoming more common in many types of work, especially through tools that can help with writing, coding, debugging, explaining code, generating documentation, customer service, and data analysis. Because of this, many students, workers, and employers are interested in how AI may change job opportunities, wages, and the skills needed in the labor market.

A useful way to study this topic is through AI exposure. In this project, AI exposure means how much an occupation's tasks may be affected, assisted, or changed by AI tools. Prior work by Eloundou et al. studies how large language models (LLMs) may affect the U.S. labor market by evaluating occupational tasks for their alignment with LLM capabilities. Their findings reveal that around 80% of the U.S. workforce could have at least 10% of their work tasks affected by the introduction of LLMs, while approximately 19% of workers may see at least 50% of their tasks impacted.


While Eloundou et al. provide a foundational framework for identifying which jobs are most exposed to generative AI, their study is inherently forward-looking and focuses strictly on theoretical task exposure. The authors themselves acknowledge a major limitation in their methodology: their exposure measure simply reflects an estimate of the technical capacity to make human labor more efficient. They note that "social, economic, regulatory, and other determinants imply that technical feasibility does not guarantee labor productivity or automation outcomes".

This distinction is what inspired our research question. Prior work effectively measures what could happen based on task characteristics, but it is missing an empirical look at what is actually happening in the real world over time. A high AI exposure score does not mean that a job will disappear. For example, software developers may have high AI exposure because AI tools can assist with coding tasks, but human workers may still be needed to design systems, review code, solve complex problems, and communicate with teams. Similarly, lower-AI-exposure jobs may still be affected indirectly by technology or the broader economy.

To fill this gap and test whether theoretical exposure translates into measurable outcomes, we plan to use data from the U.S. Bureau of Labor Statistics Occupational Employment and Wage Statistics (OEWS) program. BLS OEWS provides employment and wage estimates for many occupations, which makes it useful for comparing occupation-level labor market trends over time ("Occupational Employment and Wage Statistics Overview").

While Eloundou et al. focus specifically on LLMs, we will utilize the Artificial Intelligence Occupational Exposure (AIOE) dataset by Felten et al. as our primary empirical proxy. The AIOE provides a robust, standardized measure of how various AI capabilities link to specific occupational abilities, making it highly compatible with BLS data.

Our project will compare selected high-AI-exposure occupations with selected low-AI-exposure occupations from 2019 to 2025. This makes the project more specific than studying the entire labor market, while still allowing us to track actual employment size and wage trends before and after the rise of generative AI to see if the theoretical exposure translates to real-world impact. Since the May 2025 OEWS data are scheduled to be released in May 2026, we may include 2025 if the data are released and can be processed before the final project deadline ("Schedule of Selected Releases 2026"). If not, we will use 2019 to 2024 data and describe this limitation.

## Hypothesis

We hypothesize that occupations with above-median AI exposure, as measured by the Felten et al. AIOE index, will experience higher rates of median wage growth from 2019 to 2025 compared to occupations with below-median AI exposure. This is because many high-AI-exposure occupations involve technical, analytical, writing, coding, or information-processing tasks that often require specialized skills and may benefit from AI-assisted productivity gains.

We also hypothesize that occupations with above-median AI exposure will experience lower rates of employment growth from 2019 to 2025 compared to occupations with below-median AI exposure. This is because some tasks in high-exposure occupations may be easier for AI tools to assist, automate, or change, which could reduce the need for additional workers in some occupations.

However, we will interpret these results as trends and associations rather than causal proof. Other factors, such as COVID-19 effects, inflation, interest rates, tech layoffs, outsourcing, industry demand, and broader economic conditions, may also affect wage and employment growth during this period.

## Data

### Data overview

This project combines two occupation-level datasets. The BLS OEWS data provides employment and wage estimates, and the AIOE dataset provides occupation-level AI exposure scores. We use SOC occupation codes to merge the two datasets and compare labor market trends between above-median and below-median AI exposure occupations.


### Dataset #1: BLS Occupational Employment and Wage Statistics (OEWS), May 2023 National Estimates

* **Dataset Name:** BLS OEWS May 2023 National Estimates
* **Link to the dataset:** https://www.bls.gov/oes/tables.htm 
* **Number of observations:** 811 occupation-level observations (after cleaning and filtering).
* **Number of variables:** 5 variables relevant to our analysis.

**Description of the variables most relevant to this project:**
* **`OCC_CODE` (String):** The Standard Occupational Classification (SOC) code. This is a crucial key that uniquely identifies each occupation, which will allow us to merge this labor data with our AI exposure dataset.
* **`OCC_TITLE` (String):** The descriptive title of the occupation.
* **`TOT_EMP` (Numeric):** Estimated total employment. This represents the absolute number of workers in that occupation nationwide. It is essential for testing our hypothesis regarding employment growth rates.
* **`A_MEDIAN` (Numeric, in USD):** The median annual wage for the occupation. We use the median rather than the mean to avoid skewed distributions caused by extreme high earners. This is the primary metric for our wage trend hypothesis.
* **`A_MEAN` (Numeric, in USD):** The mean annual wage, kept as a secondary comparative metric.

**Descriptions of any shortcomings this dataset has with respect to the project:**
1. **Survey Limitations:** The BLS OEWS data is collected through employer surveys. This means it primarily captures workers in the formal economy and often excludes self-employed individuals, gig economy workers, and undocumented workers. Since many high-AI-exposure roles might involve freelancing or contract work, the `TOT_EMP` might underestimate the true size of the workforce.
2. **Lack of AI Metrics:** This dataset contains strictly labor market metrics and has no native measure of AI exposure, necessitating a merge with a secondary dataset.
3. **Changes in SOC Definitions:** While not a massive issue for a single year, as we expand this data back to 2019, we may encounter occupations where the SOC codes or definitions were updated, requiring manual alignment.

*Note for Checkpoint : For this Data Checkpoint, we are using the 2023 BLS OEWS dataset as our primary prototype to demonstrate our data wrangling, cleaning, and merging pipeline. Because the BLS data format remains largely consistent across years, successfully tidying the 2023 data proves our methodology works. We will apply this exact pipeline to concatenate the remaining years (2019-2022, 2024-2025) prior to the EDA Checkpoint.*

### Dataset #2: Artificial Intelligence Occupational Exposure (AIOE) - Language Modeling

* **Dataset Name:** AIOE - Language Modeling Scores (Felten et al., 2021)
* **Link to the dataset:** https://github.com/AIOE-Data/AIOE
* **Number of observations:** 774 occupation-level observations in the original AIOE dataset; 668 observations remain after merging with the cleaned BLS dataset.
* **Number of variables:** 3 original variables relevant to our analysis; after cleaning, we add one derived variable, ai_exposure_group.

**Description of the variables most relevant to this project:**
* **`SOC Code` (String):** The Standard Occupational Classification code. This serves as the primary key to link the AI exposure scores with the BLS labor metrics.
* **`Occupation Title` (String):** The official name of the occupation.
* **`Language Modeling AIOE` (Numeric):** A standardized z-score representing an occupation's exposure to Large Language Models (LLMs). A score of 0 represents the average exposure across all occupations; a positive score indicates above-average exposure (high potential for AI assistance/augmentation), while a negative score indicates below-average exposure.

**Descriptions of any shortcomings this dataset has with respect to the project:**
1. **Theoretical Nature:** The AIOE index measures technical exposure—the theoretical capacity for AI to perform occupational tasks—rather than actual real-world adoption or job displacement rates.
2. **Version Mismatch:** The AIOE scores are based on an older SOC taxonomy. This resulted in the loss of approximately 130 occupations during the merge process, as some newer or updated 2023 BLS job codes did not have a direct match in the AIOE dataset.
3. **Subjectivity in Mapping:** The scores are derived from mapping AI applications to human abilities via human and algorithmic labeling, which may introduce subjective bias or fail to capture the complexity of high-level human interpersonal tasks.

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

years = range(2019, 2026)
bls_yearly_dfs = []

for year in years:
    raw_path = f"data/00-raw/bls_raw_{year}.xlsx"
    if not os.path.exists(raw_path):
        print(f"  -> [Warning] Missing data file: {raw_path}. Skipping...")
        continue
    print(f"  -> Loading {year} Excel data")
    temp_df = pd.read_excel(raw_path)

    temp_df.columns = temp_df.columns.str.upper().str.strip()

    if "O_GROUP" in temp_df.columns:
        temp_df = temp_df[temp_df["O_GROUP"].str.lower() == "detailed"].copy()

    keep_cols = ["OCC_CODE", "OCC_TITLE", "TOT_EMP", "A_MEDIAN", "A_MEAN", "H_MEDIAN"]
    available_cols = [col for col in keep_cols if col in temp_df.columns]
    temp_df = temp_df[available_cols].copy()

    temp_df["A_MEDIAN_TOPCODED"] = temp_df["A_MEDIAN"].astype(str).str.strip().eq("#")
    temp_df["A_MEDIAN_SUPPRESSED"] = temp_df["A_MEDIAN"].astype(str).str.strip().eq("*")

    temp_df["YEAR"] = year
    bls_yearly_dfs.append(temp_df)

if not bls_yearly_dfs:
    raise FileNotFoundError("No BLS files found in data/00-raw/.")

n_years_loaded = len(bls_yearly_dfs)
bls_long = pd.concat(bls_yearly_dfs, ignore_index=True)
print(f"  -> Successfully concatenated {n_years_loaded} years of BLS data.\n")

bls_long["OCC_CODE"] = bls_long["OCC_CODE"].astype(str).str.strip()
bls_long["OCC_CODE"] = bls_long["OCC_CODE"].str.replace("/", "-", regex=False)
bls_long["OCC_TITLE"] = bls_long["OCC_TITLE"].astype(str).str.strip()
bls_long = bls_long.drop_duplicates(subset=["YEAR", "OCC_CODE"])

placeholder_report = (
    bls_long.groupby("YEAR")[["A_MEDIAN_TOPCODED", "A_MEDIAN_SUPPRESSED"]].sum()
)
print("Placeholder counts in A_MEDIAN by year ('#' = top-coded, '*' = suppressed):")
print(placeholder_report, "\n")

for col in ["TOT_EMP", "A_MEDIAN", "A_MEAN", "H_MEDIAN", "YEAR"]:
    if col in bls_long.columns:
        bls_long[col] = pd.to_numeric(bls_long[col], errors="coerce")

ever_topcoded = set(bls_long.loc[bls_long["A_MEDIAN_TOPCODED"], "OCC_CODE"])
bls_long["EVER_TOPCODED"] = bls_long["OCC_CODE"].isin(ever_topcoded)
print(f"Occupations top-coded in at least one year (excluded from analysis): "
      f"{len(ever_topcoded)}")
print(sorted(ever_topcoded), "\n")

bls_long = bls_long.dropna(subset=["OCC_CODE", "OCC_TITLE", "TOT_EMP", "A_MEDIAN", "YEAR"])
bls_long["YEAR"] = bls_long["YEAR"].astype(int)

years_per_occ = bls_long.groupby("OCC_CODE")["YEAR"].nunique()
balanced_occs = set(years_per_occ[years_per_occ == n_years_loaded].index)
bls_long["IN_BALANCED_PANEL"] = bls_long["OCC_CODE"].isin(balanced_occs)
print(f"Occupations with complete data in all {n_years_loaded} years "
      f"(balanced panel): {len(balanced_occs)}\n")

os.makedirs("data/02-processed", exist_ok=True)
bls_long.to_csv("data/02-processed/cleaned_bls_2019_2025.csv", index=False)

aioe_excel_path = "data/00-raw/Language Modeling AIOE and AIIE.xlsx"
if os.path.exists(aioe_excel_path):
    aioe_raw = pd.read_excel(aioe_excel_path, sheet_name="LM AIOE")
else:
    print("  -> Local Excel file not found. Fetching dynamically from GitHub repository...")
    aioe_raw = pd.read_excel(
        "https://github.com/AIOE-Data/AIOE/raw/main/Language%20Modeling%20AIOE%20and%20AIIE.xlsx",
        sheet_name="LM AIOE"
    )

aioe_clean = aioe_raw[["SOC Code", "Occupation Title", "Language Modeling AIOE"]].copy()
aioe_clean = aioe_clean.rename(columns={
    "SOC Code": "soc_code",
    "Occupation Title": "occupation_title",
    "Language Modeling AIOE": "aioe_score"
})
aioe_clean["soc_code"] = aioe_clean["soc_code"].astype(str).str.strip()
aioe_clean["occupation_title"] = aioe_clean["occupation_title"].astype(str).str.strip()
aioe_clean["aioe_score"] = pd.to_numeric(aioe_clean["aioe_score"], errors="coerce")
aioe_clean = aioe_clean.dropna(subset=["soc_code", "occupation_title", "aioe_score"])
aioe_clean = aioe_clean.drop_duplicates()

median_aioe = aioe_clean["aioe_score"].median()
aioe_clean["ai_exposure_group"] = np.where(
    aioe_clean["aioe_score"] > median_aioe, "above_median", "below_median"
)


analysis_bls = bls_long[
    bls_long["IN_BALANCED_PANEL"] & ~bls_long["EVER_TOPCODED"]
].copy()

final_df = pd.merge(
    analysis_bls,
    aioe_clean,
    left_on="OCC_CODE",
    right_on="soc_code",
    how="inner"
)
final_df = final_df.drop_duplicates(subset=["YEAR", "OCC_CODE"])
final_df = final_df.dropna(
    subset=["TOT_EMP", "A_MEDIAN", "aioe_score", "ai_exposure_group", "YEAR"]
)

ordered_cols = [
    "YEAR", "OCC_CODE", "OCC_TITLE",
    "TOT_EMP", "A_MEDIAN", "A_MEAN",
    "aioe_score", "ai_exposure_group"
]
final_df = final_df[ordered_cols]
final_df.to_csv("data/02-processed/final_analysis_dataset.csv", index=False)

print("Final merged dataset shape:", final_df.shape)
print("Years:", sorted(final_df["YEAR"].unique()))
print("Unique occupations:", final_df["OCC_CODE"].nunique())
print("Observations per year (should be identical across years):")
print(final_df["YEAR"].value_counts().sort_index())
print("Group sizes (occupations per exposure group):")
print(final_df[final_df["YEAR"] == final_df["YEAR"].max()]["ai_exposure_group"].value_counts())
display(final_df.head())
```

### 2019–2025 Longitudinal EDA Draft

The cleaned 2019–2025 BLS data lets us look at changes over time instead of only one year. Here we use the merged BLS-AIOE data to compare wage and employment patterns between above-median and below-median AI exposure occupations.


```python
wage_trend = (
    final_df
    .groupby(["YEAR", "ai_exposure_group"])["A_MEDIAN"]
    .mean()
    .reset_index()
)

plt.figure(figsize=(8, 5))
for group, group_df in wage_trend.groupby("ai_exposure_group"):
    group_df = group_df.sort_values("YEAR")
    plt.plot(group_df["YEAR"], group_df["A_MEDIAN"], marker="o", label=group)

plt.xlabel("Year")
plt.ylabel("Average Median Annual Wage")
plt.title("Average Median Wage by AI Exposure Group, 2019-2025")
plt.legend(title="AI Exposure Group")
plt.tight_layout()
plt.show()

display(wage_trend)
```

Both exposure groups show wage changes across the 2019–2025 period. This plot is useful because it moves beyond the one-year 2023 comparison and shows whether the two groups follow similar or different wage patterns over time.


```python
employment_trend = (
    final_df
    .groupby(["YEAR", "ai_exposure_group"])["TOT_EMP"]
    .sum()
    .reset_index()
    .sort_values(["ai_exposure_group", "YEAR"])
)

employment_trend["employment_index_2019"] = (
    employment_trend
    .groupby("ai_exposure_group")["TOT_EMP"]
    .transform(lambda x: x / x.iloc[0] * 100)
)

plt.figure(figsize=(8, 5))
for group, group_df in employment_trend.groupby("ai_exposure_group"):
    group_df = group_df.sort_values("YEAR")
    plt.plot(group_df["YEAR"], group_df["employment_index_2019"], marker="o", label=group)

plt.xlabel("Year")
plt.ylabel("Employment Index, 2019 = 100")
plt.title("Employment Index by AI Exposure Group, 2019-2025")
plt.legend(title="AI Exposure Group")
plt.tight_layout()
plt.show()

display(employment_trend)
```

The employment trend is shown as an index with 2019 set to 100. This makes the two exposure groups easier to compare because it focuses on relative change instead of raw employment size.


```python
growth_source = final_df[final_df["YEAR"].isin([2019, 2025])].copy()

growth_wide = growth_source.pivot_table(
    index=["OCC_CODE", "OCC_TITLE", "ai_exposure_group", "aioe_score"],
    columns="YEAR",
    values=["A_MEDIAN", "TOT_EMP"],
    aggfunc="first"
)

growth_wide.columns = [f"{var}_{year}" for var, year in growth_wide.columns]
growth_df = growth_wide.reset_index()
growth_df = growth_df.dropna(subset=["A_MEDIAN_2019", "A_MEDIAN_2025", "TOT_EMP_2019", "TOT_EMP_2025"])

growth_df["wage_growth_pct"] = (
    (growth_df["A_MEDIAN_2025"] - growth_df["A_MEDIAN_2019"])
    / growth_df["A_MEDIAN_2019"]
    * 100
)

growth_df["employment_growth_pct"] = (
    (growth_df["TOT_EMP_2025"] - growth_df["TOT_EMP_2019"])
    / growth_df["TOT_EMP_2019"]
    * 100
)

growth_df = growth_df.replace([np.inf, -np.inf], np.nan)
growth_df = growth_df.dropna(subset=["wage_growth_pct", "employment_growth_pct"])

growth_summary = growth_df.groupby("ai_exposure_group")[[
    "wage_growth_pct",
    "employment_growth_pct"
]].agg(["count", "mean", "median"]).round(2)

display(growth_summary)
```

```python
plt.figure(figsize=(8, 5))
growth_df.boxplot(column="wage_growth_pct", by="ai_exposure_group")
plt.title("2019-2025 Wage Growth by AI Exposure Group")
plt.suptitle("")
plt.xlabel("AI Exposure Group")
plt.ylabel("Wage Growth (%)")
plt.tight_layout()
plt.show()

plt.figure(figsize=(8, 5))
growth_df.boxplot(column="employment_growth_pct", by="ai_exposure_group")
plt.title("2019-2025 Employment Growth by AI Exposure Group")
plt.suptitle("")
plt.xlabel("AI Exposure Group")
plt.ylabel("Employment Growth (%)")
plt.tight_layout()
plt.show()
```

These growth plots compare occupations that appear in both 2019 and 2025. Wage growth is closer to the first part of our hypothesis, while employment growth is closer to the second part. The final analysis can build on this by adding a t-test or regression.

## Ethics

This project has relatively low privacy risk because we plan to use public occupation-level data, not individual-level data. We will not collect personal names, resumes, individual salaries, private employment records, or private company data.

The main ethical concern is interpretation. AI exposure should not be treated as the same thing as job loss risk. A high AI exposure score means that some tasks in an occupation may be affected or assisted by AI, but it does not mean the occupation will disappear. We will be careful to describe our results as trends and associations, not causal proof.

There may also be bias in the data or in the way AI exposure is measured. National-level occupation averages may hide differences by location, company size, worker experience, education level, and specific skills. Labor market trends may also be affected by factors other than AI, such as COVID-19 effects, interest rates, tech layoffs, outsourcing, industry demand, and broader economic conditions. We will discuss these limitations when communicating our results.
