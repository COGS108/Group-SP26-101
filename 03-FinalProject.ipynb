## Abstract

Artificial intelligence is widely predicted to reshape work, but most existing measures of AI exposure describe what AI *could* affect rather than what is actually happening in the labor market. This project tests whether theoretical exposure has translated into measurable outcomes: among U.S. occupations, do those with above-median AI exposure — measured by the Felten et al. Language Modeling AIOE index — experience higher median wage growth and lower employment growth from 2019 to 2025 than less-exposed occupations? We merge seven years of BLS Occupational Employment and Wage Statistics with the AIOE scores, building a balanced panel of 668 occupations with harmonized SOC codes, and compare exposure groups using Welch's t-tests, Mann-Whitney U tests, OLS regressions with baseline controls, and a pre/post-generative-AI sub-period analysis. Neither hypothesis is supported. Low-exposure occupations saw higher nominal wage growth (28.2% vs 21.5%, p < 0.001), but regression analysis shows this is largely a low-wage base effect from pandemic-era raises rather than an AI effect. High-exposure occupations grew in employment (+4.2% vs −6.4%, p < 0.001), an advantage that predates ChatGPT. Through May 2025, theoretical AI exposure has not translated into aggregate displacement; these results are associations over a turbulent period, not causal estimates.

## Research Question

Among U.S. occupations, do those with above-median AI exposure, as measured by the Felten et al. AIOE index, experience higher rates of median wage growth and lower rates of employment growth from 2019 to 2025 compared to occupations with below-median AI exposure?

## Background and Prior Work

Artificial intelligence is becoming more common in many types of work, especially through tools that can help with writing, coding, debugging, explaining code, generating documentation, customer service, and data analysis. Because of this, many students, workers, and employers are interested in how AI may change job opportunities, wages, and the skills needed in the labor market.

A useful way to study this topic is through AI exposure. In this project, AI exposure means how much an occupation's tasks may be affected, assisted, or changed by AI tools. Prior work by Eloundou et al. studies how large language models (LLMs) may affect the U.S. labor market by evaluating occupational tasks for their alignment with LLM capabilities. Their findings reveal that around 80% of the U.S. workforce could have at least 10% of their work tasks affected by the introduction of LLMs, while approximately 19% of workers may see at least 50% of their tasks impacted.


While Eloundou et al. provide a foundational framework for identifying which jobs are most exposed to generative AI, their study is inherently forward-looking and focuses strictly on theoretical task exposure. The authors themselves acknowledge a major limitation in their methodology: their exposure measure simply reflects an estimate of the technical capacity to make human labor more efficient. They note that "social, economic, regulatory, and other determinants imply that technical feasibility does not guarantee labor productivity or automation outcomes".

This distinction is what inspired our research question. Prior work effectively measures what could happen based on task characteristics, but it is missing an empirical look at what is actually happening in the real world over time. A high AI exposure score does not mean that a job will disappear. For example, software developers may have high AI exposure because AI tools can assist with coding tasks, but human workers may still be needed to design systems, review code, solve complex problems, and communicate with teams. Similarly, lower-AI-exposure jobs may still be affected indirectly by technology or the broader economy.

To fill this gap and test whether theoretical exposure translates into measurable outcomes, we plan to use data from the U.S. Bureau of Labor Statistics Occupational Employment and Wage Statistics (OEWS) program. BLS OEWS provides employment and wage estimates for many occupations, which makes it useful for comparing occupation-level labor market trends over time ("Occupational Employment and Wage Statistics Overview").

While Eloundou et al. focus specifically on LLMs, we will utilize the Artificial Intelligence Occupational Exposure (AIOE) dataset by Felten et al. as our primary empirical proxy. The AIOE provides a robust, standardized measure of how various AI capabilities link to specific occupational abilities, making it highly compatible with BLS data.

Our project will compare selected high-AI-exposure occupations with selected low-AI-exposure occupations from 2019 to 2025. This makes the project more specific than studying the entire labor market, while still allowing us to track actual employment size and wage trends before and after the rise of generative AI to see if the theoretical exposure translates to real-world impact.

## Hypothesis

We hypothesize that occupations with above-median AI exposure, as measured by the Felten et al. AIOE index, will experience higher rates of median wage growth from 2019 to 2025 compared to occupations with below-median AI exposure. This is because many high-AI-exposure occupations involve technical, analytical, writing, coding, or information-processing tasks that often require specialized skills and may benefit from AI-assisted productivity gains.

We also hypothesize that occupations with above-median AI exposure will experience lower rates of employment growth from 2019 to 2025 compared to occupations with below-median AI exposure. This is because some tasks in high-exposure occupations may be easier for AI tools to assist, automate, or change, which could reduce the need for additional workers in some occupations.

However, we will interpret these results as trends and associations rather than causal proof. Other factors, such as COVID-19 effects, inflation, interest rates, tech layoffs, outsourcing, industry demand, and broader economic conditions, may also affect wage and employment growth during this period.

## Data

### Data overview

This project combines two occupation-level datasets. The BLS OEWS data provides employment and wage estimates, and the AIOE dataset provides occupation-level AI exposure scores. We use SOC occupation codes to merge the two datasets and compare labor market trends between above-median and below-median AI exposure occupations.


### Dataset #1: BLS Occupational Employment and Wage Statistics (OEWS), National Estimates, May 2019 – May 2025

* **Dataset Name:** BLS OEWS National Estimates, May 2019 – May 2025 
* **Link to the dataset:** https://www.bls.gov/oes/tables.htm
* **Number of observations:** Each annual file contains 789–831 detailed (6-digit SOC) occupations after removing total/major/minor/broad aggregation rows (789 in 2019–2020 under the hybrid SOC; 830–831 in 2021–2025 under the 2018 SOC). After cleaning, concatenating, and SOC harmonization, 748 occupations have valid wage and employment data in all seven years (the balanced panel); excluding 21 ever-top-coded occupations, 668 of these merge with the AIOE data to form the final analysis sample of 4,676 occupation-year observations (668 × 7).
* **Number of variables:** 6 raw variables retained from each file, plus derived variables created during cleaning (`YEAR`, placeholder flags, and panel-membership flags).

**Description of the variables most relevant to this project:**

* **`OCC_CODE` (String):** The Standard Occupational Classification (SOC) code. This is a crucial key that uniquely identifies each occupation, which allows us to merge this labor data with our AI exposure dataset.
* **`OCC_TITLE` (String):** The descriptive title of the occupation.
* **`TOT_EMP` (Numeric):** Estimated total employment. This represents the absolute number of workers in that occupation nationwide. It is essential for testing our hypothesis regarding employment growth rates.
* **`A_MEDIAN` (Numeric, in USD):** The median annual wage for the occupation. We use the median rather than the mean to avoid skewed distributions caused by extreme high earners. This is the primary metric for our wage trend hypothesis.
* **`A_MEAN` (Numeric, in USD):** The mean annual wage, kept as a secondary comparative metric.
* **`YEAR` (Numeric, derived):** The OEWS reference year (May 2019 – May 2025), added when concatenating the seven annual files into a long-format panel.
* **`A_MEDIAN_TOPCODED` / `A_MEDIAN_SUPPRESSED` (Boolean, derived):** Flags marking the BLS placeholder symbols in the raw wage column (`#` = wage above the top-code threshold; `*` = estimate not available), used to filter the analysis sample transparently.

**Descriptions of any shortcomings this dataset has with respect to the project:**

1. **Survey Limitations:** The BLS OEWS data is collected through employer surveys. This means it primarily captures workers in the formal economy and often excludes self-employed individuals, gig economy workers, and undocumented workers. Since many high-AI-exposure roles might involve freelancing or contract work, `TOT_EMP` might underestimate the true size of the workforce.
2. **Lack of AI Metrics:** This dataset contains strictly labor market metrics and has no native measure of AI exposure, necessitating a merge with a secondary dataset.
3. **Changes in SOC Definitions:** BLS switched from a hybrid SOC taxonomy to the full 2018 SOC in May 2021. As a result, three computer-occupation groups (software developers, web developers, and database administrators) are reported as combined codes in 2019–2020 but as split codes in 2021–2025. We harmonize toward the combined codes by aggregating the split codes: employment is summed, and the combined median wage is approximated by the employment-weighted average of the component medians (the true pooled median is not recoverable from published aggregates). Without this step, roughly 1.9 million software-related workers — occupations central to an AI-exposure study — would silently drop out of the balanced panel.
4. **Top-coding and Suppression:** Annual wages above the BLS top-code threshold are reported only as `#`, and some estimates are suppressed as `*`. Top-coding is also inconsistent across years: several physician and dentist occupations are top-coded in 2019–2024 but reported with actual values in 2025, which would create an artificial jump in 2025 group averages if kept. We therefore exclude all 21 occupations that are top-coded in any year, at the cost of removing some of the highest-wage occupations from the analysis.


### Dataset #2: Artificial Intelligence Occupational Exposure (AIOE) - Language Modeling

* **Dataset Name:** AIOE - Language Modeling Scores (Felten et al., 2021)
* **Link to the dataset:** https://github.com/AIOE-Data/AIOE
* **Number of observations:** 774 occupation-level observations in the original AIOE dataset, plus 12 crosswalk-constructed observations for 2018-SOC computer occupations (described below); 668 occupations remain in the final merged analysis sample.
* **Number of variables:** 3 original variables relevant to our analysis; after cleaning, we add one derived variable, `ai_exposure_group`.

**Description of the variables most relevant to this project:**

* **`SOC Code` (String):** The Standard Occupational Classification code. This serves as the primary key to link the AI exposure scores with the BLS labor metrics.
* **`Occupation Title` (String):** The official name of the occupation.
* **`Language Modeling AIOE` (Numeric):** A standardized z-score representing an occupation's exposure to Large Language Models (LLMs). A score of 0 represents the average exposure across all occupations; a positive score indicates above-average exposure (high potential for AI assistance/augmentation), while a negative score indicates below-average exposure.
* **`ai_exposure_group` (String, derived):** "above_median" or "below_median", defined by splitting occupations at the median of the original published AIOE distribution (computed before any crosswalk additions, so the threshold is not affected by our own data processing).

**Descriptions of any shortcomings this dataset has with respect to the project:**

1. **Theoretical Nature:** The AIOE index measures technical exposure—the theoretical capacity for AI to perform occupational tasks—rather than actual real-world adoption or job displacement rates. It was also constructed before the most recent generation of AI models, so it reflects the ability–AI links visible at the time of its creation.
2. **Version Mismatch:** The AIOE scores use the 2010 SOC taxonomy, while BLS renumbered the entire computer-occupation group (15-12xx) under the 2018 SOC. Without intervention, the occupations most central to an AI-exposure study — software developers, programmers, systems analysts — would drop out of the merge entirely. We therefore built a manual 2010→2018 crosswalk for 12 computer-occupation codes, assigning the mean of the source scores where one 2018 code corresponds to multiple 2010 codes. Even with the crosswalk, 106 of the 774 original AIOE occupations do not appear in the final sample, due to a combination of remaining unmatched SOC codes and our balanced-panel and top-code filters on the BLS side.
3. **Subjectivity in Mapping:** The scores are derived from mapping AI applications to human abilities via human and algorithmic labeling, which may introduce subjective bias or fail to capture the complexity of high-level human interpersonal tasks.

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#  Load and concatenate yearly BLS files 

years = range(2019, 2026)
bls_yearly_dfs = []

for year in years:
    raw_path = "data/00-raw/bls_raw_" + str(year) + ".xlsx"
    if not os.path.exists(raw_path):
        print("  -> [Warning] Missing data file:", raw_path, ". Skipping...")
        continue
    print("  -> Loading", year, "Excel data")
    temp_df = pd.read_excel(raw_path)

    # 2019 file uses lowercase column names; standardize to uppercase
    temp_df.columns = temp_df.columns.str.upper().str.strip()

    # Keep only detailed (6-digit SOC) occupations. Raw files also contain
    # total/major/minor/broad aggregation rows, and a few codes appear twice 
    # (as both "broad" and "detailed"),  which would double-count if kept.
    if "O_GROUP" in temp_df.columns:
        temp_df = temp_df[temp_df["O_GROUP"].str.lower() == "detailed"].copy()

    keep_cols = ["OCC_CODE", "OCC_TITLE", "TOT_EMP", "A_MEDIAN", "A_MEAN", "H_MEDIAN"]
    available_cols = [col for col in keep_cols if col in temp_df.columns]
    temp_df = temp_df[available_cols].copy()

    #   '#' = annual wage above the BLS top-code threshold ($208,000 in 2019-2023; $239,200 in 2024). NOT missing data.
    #   '*' = estimate not available (hourly-only occupations such as Actors, Musicians, Dancers).
    
    # In 2019-2024 files, physician/dentist occupations are reported as '#',
    # but the 2025 file reports their actual values. Without handling this,
    # these very high-wage occupations enter the sample ONLY in 2025,
    # creating an artificial jump in the 2025 group averages.
    
    temp_df["A_MEDIAN_TOPCODED"] = temp_df["A_MEDIAN"].astype(str).str.strip().eq("#")
    temp_df["A_MEDIAN_SUPPRESSED"] = temp_df["A_MEDIAN"].astype(str).str.strip().eq("*")

    temp_df["YEAR"] = year
    bls_yearly_dfs.append(temp_df)

if not bls_yearly_dfs:
    raise FileNotFoundError("No BLS files found in data/00-raw/.")

n_years_loaded = len(bls_yearly_dfs)
bls_long = pd.concat(bls_yearly_dfs, ignore_index=True)

print("  -> Successfully concatenated", n_years_loaded, "years of BLS data.\n")


# Clean types and document placeholder handling
bls_long["OCC_CODE"] = bls_long["OCC_CODE"].astype(str).str.strip()
n_repaired = bls_long["OCC_CODE"].str.match(r"^\d{2}-\d{4}$").sum()
assert n_repaired == len(bls_long), "Some OCC_CODE values are still not in 'MM-NNNN' format!"
bls_long["OCC_TITLE"] = bls_long["OCC_TITLE"].astype(str).str.strip()
bls_long = bls_long.drop_duplicates(subset=["YEAR", "OCC_CODE"])

# report how many placeholder cells exist per year
placeholder_report = (
    bls_long.groupby("YEAR")[["A_MEDIAN_TOPCODED", "A_MEDIAN_SUPPRESSED"]].sum()
)
print("Placeholder counts in A_MEDIAN by year ('#' = top-coded, '*' = suppressed):")
print(placeholder_report, "\n")

for col in ["TOT_EMP", "A_MEDIAN", "A_MEAN", "H_MEDIAN", "YEAR"]:
    if col in bls_long.columns:
        bls_long[col] = pd.to_numeric(bls_long[col], errors="coerce")

# if an occupation's median wage was top-coded in ANY year, 
# exclude it from ALL years. Otherwise it would be present in some years and absent in others, 
# biasing year-to-year trends.

ever_topcoded = set(bls_long.loc[bls_long["A_MEDIAN_TOPCODED"], "OCC_CODE"])
bls_long["EVER_TOPCODED"] = bls_long["OCC_CODE"].isin(ever_topcoded)

print("Occupations top-coded in at least one year (excluded from analysis):",
      len(ever_topcoded))
print(sorted(ever_topcoded), "\n")

bls_long = bls_long.dropna(subset=["OCC_CODE", "OCC_TITLE", "TOT_EMP", "A_MEDIAN", "YEAR"])
bls_long["YEAR"] = bls_long["YEAR"].astype(int)


# BLS switched from the hybrid SOC to the full 2018 SOC in May 2021. 
# Three computer-occupation groups were reported as COMBINED codes in 2019-2020
# but as SPLIT codes in 2021-2025,
# so none of them survives the balanced-panel filter and ~1.9M software
# workers silently vanish. We harmonize toward the COMBINED codes by
# aggregating the split codes (aggregation is clean; splitting wages is not):
#   TOT_EMP  -> sum of components
#   A_MEDIAN -> employment-weighted average of component medians
#               (an approximation of the combined group's median; the true
#               pooled median is not recoverable from published aggregates)


SOC_HARMONIZATION = {
    "15-1256": (["15-1252", "15-1253"],
                "Software Developers and Software Quality Assurance Analysts and Testers"),
    "15-1257": (["15-1254", "15-1255"],
                "Web Developers and Digital Interface Designers"),
    "15-1245": (["15-1242", "15-1243"],
                "Database Administrators and Architects"),
}

harmonized_rows = []
for target_code, (split_codes, target_title) in SOC_HARMONIZATION.items():
    parts = bls_long[bls_long["OCC_CODE"].isin(split_codes)]
    for year, grp in parts.groupby("YEAR"):
        if len(grp) < len(split_codes):
            continue  
        w = grp["TOT_EMP"]
        harmonized_rows.append({
            "YEAR": year,
            "OCC_CODE": target_code,
            "OCC_TITLE": target_title,
            "TOT_EMP": w.sum(),
            "A_MEDIAN": np.average(grp["A_MEDIAN"], weights=w),
            "A_MEAN": np.average(grp["A_MEAN"], weights=w),
            "H_MEDIAN": np.average(grp["H_MEDIAN"], weights=w),
            "A_MEDIAN_TOPCODED": grp["A_MEDIAN_TOPCODED"].any(),
            "A_MEDIAN_SUPPRESSED": grp["A_MEDIAN_SUPPRESSED"].any(),
            "EVER_TOPCODED": grp["EVER_TOPCODED"].any(),
        })

all_split_codes = [c for codes, _ in SOC_HARMONIZATION.values() for c in codes]
bls_long = bls_long[~bls_long["OCC_CODE"].isin(all_split_codes)]
bls_long = pd.concat([bls_long, pd.DataFrame(harmonized_rows)], ignore_index=True)

print("SOC harmonization: aggregated", len(harmonized_rows),
      "year-rows of split codes back into", len(SOC_HARMONIZATION),
      "combined occupations (" + ", ".join(SOC_HARMONIZATION) + ").\n")

# keep only occupations observed with valid wage and employment data in EVERY loaded year,
# so that yearly group averages compare the same set of occupations instead of mixing 
# in entry/exit composition effects. 

years_per_occ = bls_long.groupby("OCC_CODE")["YEAR"].nunique()
balanced_occs = set(years_per_occ[years_per_occ == n_years_loaded].index)
bls_long["IN_BALANCED_PANEL"] = bls_long["OCC_CODE"].isin(balanced_occs)

print("Occupations with complete data in all", n_years_loaded,
      "years (balanced panel):", len(balanced_occs), "\n")

os.makedirs("data/02-processed", exist_ok=True)
bls_long.to_csv("data/02-processed/cleaned_bls_2019_2025.csv", index=False)

# Load and clean the AIOE dataset

aioe_excel_path = "data/00-raw/Language Modeling AIOE and AIIE.xlsx"
if os.path.exists(aioe_excel_path):
    aioe_raw = pd.read_excel(aioe_excel_path, sheet_name="LM AIOE")
else:
    print("  -> Local Excel file not found. Fetching dynamically from GitHub repository...")
    aioe_raw = pd.read_excel(
        "https://github.com/AIOE-Data/AIOE/raw/main/Language%20Modeling%20AIOE%20and%20AIIE.xlsx",
        sheet_name="LM AIOE"
    )

aioe_clean = aioe_raw[["SOC Code", "Occupation Title", "Language Modeling AIOE"]].copy()
aioe_clean = aioe_clean.rename(columns={
    "SOC Code": "soc_code",
    "Occupation Title": "occupation_title",
    "Language Modeling AIOE": "aioe_score"
})
aioe_clean["soc_code"] = aioe_clean["soc_code"].astype(str).str.strip()
aioe_clean["occupation_title"] = aioe_clean["occupation_title"].astype(str).str.strip()
aioe_clean["aioe_score"] = pd.to_numeric(aioe_clean["aioe_score"], errors="coerce")
aioe_clean = aioe_clean.dropna(subset=["soc_code", "occupation_title", "aioe_score"])
aioe_clean = aioe_clean.drop_duplicates()

median_aioe = aioe_clean["aioe_score"].median()

# the AIOE dataset uses 2010 SOC codes, but BLS renumbered the entire computer occupation group (15-11xx -> 15-12xx) in the 2018 SOC. 
# Without this mapping, programmers, software developers, systems analysts. 
# The occupations most central to an AI-exposure study -- drop out of the inner merge.

# Keys = 2018/harmonized BLS codes; values = source 2010 AIOE codes.

# Where one 2018 code corresponds to multiple 2010 codes, we assign the mean of the component AIOE scores.
# The median threshold above is computed on the ORIGINAL published AIOE distribution, before these mapped rows are appended.

AIOE_CROSSWALK_2010_TO_2018 = {
    "15-1221": ["15-1111"],             # Computer and Info Research Scientists
    "15-1211": ["15-1121"],             # Computer Systems Analysts
    "15-1212": ["15-1122"],             # Information Security Analysts
    "15-1251": ["15-1131"],             # Computer Programmers
    "15-1256": ["15-1132", "15-1133"],  # Software Developers (Apps + Systems)
    "15-1257": ["15-1134"],             # Web Developers
    "15-1245": ["15-1141"],             # Database Administrators and Architects
    "15-1244": ["15-1142"],             # Network and Computer Systems Admins
    "15-1241": ["15-1143"],             # Computer Network Architects
    "15-1232": ["15-1151"],             # Computer User Support Specialists
    "15-1231": ["15-1152"],             # Computer Network Support Specialists
    "15-1299": ["15-1199"],             # Computer Occupations, All Other
}

score_by_code = aioe_clean.set_index("soc_code")["aioe_score"]
title_by_code = aioe_clean.set_index("soc_code")["occupation_title"]
crosswalk_rows = []
for new_code, old_codes in AIOE_CROSSWALK_2010_TO_2018.items():
    
    missing = [c for c in old_codes if c not in score_by_code.index]
    assert not missing, "Crosswalk source codes not found in AIOE: " + str(missing)
    crosswalk_rows.append({
        "soc_code": new_code,
        "occupation_title": " / ".join(title_by_code[c] for c in old_codes),
        "aioe_score": float(np.mean([score_by_code[c] for c in old_codes])),
    })
    
aioe_clean = pd.concat([aioe_clean, pd.DataFrame(crosswalk_rows)], ignore_index=True)
aioe_clean = aioe_clean.drop_duplicates(subset=["soc_code"])

print("AIOE crosswalk: added", len(crosswalk_rows),
      "mapped 2018-SOC computer occupations to the AIOE table.\n")

aioe_clean["ai_exposure_group"] = np.where(
    aioe_clean["aioe_score"] > median_aioe, "above_median", "below_median"
)

# Merge and build the final analysis dataset

# Analysis sample = balanced panel, excluding ever-top-coded occupations,
# so every occupation contributes one observation per year with a real
# (non-imputed, non-censored) median wage.
analysis_bls = bls_long[
    bls_long["IN_BALANCED_PANEL"] & ~bls_long["EVER_TOPCODED"]
].copy()

final_df = pd.merge(
    analysis_bls,
    aioe_clean,
    left_on="OCC_CODE",
    right_on="soc_code",
    how="inner"
)
final_df = final_df.drop_duplicates(subset=["YEAR", "OCC_CODE"])
final_df = final_df.dropna(
    subset=["TOT_EMP", "A_MEDIAN", "aioe_score", "ai_exposure_group", "YEAR"]
)

ordered_cols = [
    "YEAR", "OCC_CODE", "OCC_TITLE",
    "TOT_EMP", "A_MEDIAN", "A_MEAN",
    "aioe_score", "ai_exposure_group"
]
final_df = final_df[ordered_cols]
final_df.to_csv("data/02-processed/final_analysis_dataset.csv", index=False)

print("Final merged dataset shape:", final_df.shape)
print("Years:", sorted(final_df["YEAR"].unique()))
print("Unique occupations:", final_df["OCC_CODE"].nunique())
print("Observations per year (should be identical across years):")
print(final_df["YEAR"].value_counts().sort_index())
print("Group sizes (occupations per exposure group):")
print(final_df[final_df["YEAR"] == final_df["YEAR"].max()]["ai_exposure_group"].value_counts())
display(final_df.head())
```

### Statistical Analysis



### Analysis 1: Welch's t-test on 2019–2025 growth rates

To formally test our two hypotheses, we first compare the mean 2019–2025 growth rates between the above-median and below-median AI exposure groups using Welch's t-test, which does not assume equal variances between groups. The unit of analysis is the occupation (n = 668), where each occupation contributes one wage growth rate and one employment growth rate.

We also report Cohen's d to quantify the size of each difference, since with samples this large even trivial differences can be statistically significant. Both tests are highly significant but in the **opposite direction of our hypotheses**: the below-median group experienced higher mean nominal wage growth (28.2% vs 21.5%, p < 0.001, d = −0.67), while the above-median group experienced higher mean employment growth (+4.2% vs −6.4%, p < 0.001, d = 0.42).

```python
from scipy import stats

def cohens_d(x, y):
    
    """Pooled-SD standardized mean difference between two samples."""
    
    nx = len(x)
    ny = len(y)
    
    pooled_var = (
        ((nx - 1) * x.var(ddof=1) + (ny - 1) * y.var(ddof=1))
        / (nx + ny - 2)
    )
    
    return (x.mean() - y.mean()) / np.sqrt(pooled_var)

above = growth_df[growth_df["ai_exposure_group"] == "above_median"]
below = growth_df[growth_df["ai_exposure_group"] == "below_median"]
 
print("Analysis 1: Welch's t-tests (above-median vs below-median exposure)")

for var, label in [("wage_growth_pct", "Wage growth"),("employment_growth_pct", "Employment growth")]:
    
    t_stat, p_val = stats.ttest_ind(above[var], below[var], equal_var=False)
    
    d = cohens_d(above[var], below[var])
    
    print("")
    print(label, "2019-2025 (%):")
    print("  above-median mean =", round(above[var].mean(), 2),
          "  below-median mean =", round(below[var].mean(), 2))
    print("  Welch t =", round(t_stat, 3),
          "  p =", p_val,
          "  Cohen's d =", round(d, 3))
```

### Analysis 2: Mann-Whitney U test (non-parametric robustness check)

The growth distributions are visibly skewed with extreme outliers (wage growth skewness reaches 2.4 in the above-median group, and employment growth skewness reaches 2.0 in the below-median group), which can violate the normality assumptions behind the t-test. We therefore repeat both comparisons with the Mann-Whitney U test.

The conclusions are unchanged. Strictly speaking, the Mann-Whitney U test compares the rank ordering of the two distributions — whether values from one group tend to exceed values from the other — rather than testing medians directly. By this criterion, wage growth tends to be higher in the below-median exposure group (descriptive medians: 28.0% vs 20.9%; U = 30,647, p < 0.001), while employment growth tends to be higher in the above-median group (descriptive medians: +2.6% vs −6.8%; U = 70,778, p < 0.001). Because the parametric and non-parametric tests agree, the group differences are unlikely to be artifacts of outliers or skew.

```python
print("Analysis 2: Mann-Whitney U tests (robustness to skew/outliers)")

for var, label in [("wage_growth_pct", "Wage growth"), ("employment_growth_pct", "Employment growth")]:
    
    u_stat, p_val = stats.mannwhitneyu(above[var], below[var], alternative="two-sided")
    
    print("")
    print(label, ": skewness above =", round(stats.skew(above[var]), 2), ", below =", round(stats.skew(below[var]), 2))
    print("  above-median median =", round(above[var].median(), 2),"  below-median median =", round(below[var].median(), 2))
    print("  Mann-Whitney U =", round(u_stat),"  p =", p_val)
```

### Analysis 3: OLS regression with continuous exposure and baseline controls

The group comparison above may be confounded by baseline wage levels: the two groups start from very different 2019 wages (about 69,500 vs 43,700 USD on average), and low-wage occupations received large pandemic-era nominal raises for reasons unrelated to AI (minimum wage increases, labor shortages in service sectors). 

To separate the role of AI exposure from this base-wage effect, we regress each growth rate on the continuous AIOE score while controlling for the log of the occupation's 2019 baseline level. The results reframe our first finding: once baseline wage is held constant, the AIOE score retains only a small negative association with wage growth (coefficient ≈ −0.9 percentage points per unit of AIOE, t = −2.33, p ≈ 0.02), whereas the baseline-wage control is a much stronger predictor within the same model (t = −11.61, p < 0.001); together the two predictors explain about 26% of the variance in wage growth (model R² = 0.259). In other words, "low-exposure occupations grew faster" is largely a low-wage base effect rather than an AI effect. For employment, the continuous result confirms the group test: a one-unit increase in AIOE score is associated with roughly +5 percentage points of 2019–2025 employment growth (t = 5.34, p < 0.001), even after controlling for 2019 employment size (model R² = 0.077).

```python
import statsmodels.formula.api as smf
 
wage_model = smf.ols(
    
    "wage_growth_pct ~ aioe_score + log_wage_2019", data=growth_df).fit()

print("Analysis 3a: wage_growth_pct ~ aioe_score + log_wage_2019")
print(wage_model.summary().tables[1])
print("R-squared =", round(wage_model.rsquared, 3))
print("")
 
emp_model = smf.ols( "employment_growth_pct ~ aioe_score + log_emp_2019", data=growth_df).fit()

print("Analysis 3b: employment_growth_pct ~ aioe_score + log_emp_2019")
print(emp_model.summary().tables[1])
print("R-squared =", round(emp_model.rsquared, 3))

fig, axes = plt.subplots(1, 2, figsize=(13, 5))

colors = {"above_median": "tab:blue", "below_median": "tab:orange"}

for grp, sub in growth_df.groupby("ai_exposure_group"):
    
    axes[0].scatter(sub["aioe_score"], sub["wage_growth_pct"], alpha=0.45, s=18, color=colors[grp], label=grp)
    axes[1].scatter(sub["aioe_score"], sub["employment_growth_pct"],alpha=0.45, s=18, color=colors[grp], label=grp)
 
xs = np.linspace(growth_df["aioe_score"].min(), growth_df["aioe_score"].max(), 100)

pred_wage = (
    wage_model.params["Intercept"]
    + wage_model.params["aioe_score"] * xs
    + wage_model.params["log_wage_2019"] * growth_df["log_wage_2019"].mean()
)
pred_emp = (
    emp_model.params["Intercept"]
    + emp_model.params["aioe_score"] * xs
    + emp_model.params["log_emp_2019"] * growth_df["log_emp_2019"].mean()
)

axes[0].plot(xs, pred_wage, color="black", lw=2,label="OLS fit (at mean 2019 wage)")
axes[1].plot(xs, pred_emp, color="black", lw=2, label="OLS fit (at mean 2019 employment)")
 
axes[0].set_xlabel("AIOE Score")
axes[0].set_ylabel("2019-2025 Wage Growth (%)")
axes[0].set_title("AI Exposure vs Wage Growth, 2019-2025")
axes[0].legend(fontsize=8)

axes[1].set_xlabel("AIOE Score")
axes[1].set_ylabel("2019-2025 Employment Growth (%)")
axes[1].set_title("AI Exposure vs Employment Growth, 2019-2025")
axes[1].legend(fontsize=8)

plt.tight_layout()
plt.show()
```

### Analysis 4: Pre- vs post-generative-AI sub-periods

Our six-year window mixes two very different shocks: the COVID-19 pandemic (2020–2021) and the rise of generative AI after ChatGPT's release in November 2022. To check whether the patterns above are driven by the generative-AI era specifically, we split the window into two three-year sub-periods, 2019–2022 and 2022–2025, and repeat the group comparison within each. The wage growth gap favoring the below-median group narrows substantially after 2022 (from −4.7 to −1.5 percentage points, both p < 0.01), consistent with the pandemic-era low-wage boom fading. The employment advantage of high-exposure occupations appears in **both** sub-periods (+6.5 and +3.8 percentage points), suggesting it predates generative AI and may reflect longer-run demand for analytical and information-processing work rather than an AI-specific effect. We interpret these sub-period results cautiously, as three-year windows are short and other macroeconomic forces changed across the same boundary.

```python
def period_growth(df, y0, y1):
    
    """Per-occupation % growth in median wage and employment
    between year y0 and year y1, with exposure-group labels."""
    
    sub = df[df["YEAR"].isin([y0, y1])]
    
    wide = sub.pivot_table(
        index=["OCC_CODE", "ai_exposure_group"],
        columns="YEAR",
        values=["A_MEDIAN", "TOT_EMP"],
        aggfunc="first",
    ).dropna()
    
    wide.columns = [str(v) + "_" + str(y) for v, y in wide.columns]
    
    wide = wide.reset_index()
    
    start_wage = "A_MEDIAN_" + str(y0)
    end_wage = "A_MEDIAN_" + str(y1)
    start_emp = "TOT_EMP_" + str(y0)
    end_emp = "TOT_EMP_" + str(y1)
    
    wide["wage_growth_pct"] = (
        (wide[end_wage] - wide[start_wage]) / wide[start_wage] * 100
    )
    
    wide["employment_growth_pct"] = (
        (wide[end_emp] - wide[start_emp]) / wide[start_emp] * 100
    )
    
    return wide
 
 
print("Analysis 4: pre-generative-AI (2019-2022) vs post (2022-2025)")

period_results = []

for y0, y1, label in [(2019, 2022, "2019-2022 (pre)"), (2022, 2025, "2022-2025 (post)")]:
    pg = period_growth(final_df, y0, y1)
    
    a = pg[pg["ai_exposure_group"] == "above_median"]
    b = pg[pg["ai_exposure_group"] == "below_median"]
    
    for var, vlabel in [("wage_growth_pct", "wage"), ("employment_growth_pct", "employment")]:
        
        t_stat, p_val = stats.ttest_ind(a[var], b[var], equal_var=False)
        
        period_results.append({
            "period": label,
            "outcome": vlabel,
            "above_mean": round(a[var].mean(), 2),
            "below_mean": round(b[var].mean(), 2),
            "gap (above - below)": round(a[var].mean() - b[var].mean(), 2),
            "welch_p": p_val,
        })
        
period_table = pd.DataFrame(period_results)
display(period_table)
```

## Discussion and Conclusion



### Summary of results

This project asked whether U.S. occupations with above-median AI exposure, as measured by the Felten et al. AIOE index, experienced higher median wage growth and lower employment growth from 2019 to 2025 than occupations with below-median exposure. Using a balanced panel of 668 occupations built from BLS OEWS data, we found that the data supported **neither** hypothesis — in fact, both raw comparisons ran in the opposite direction.

For wages, the below-median group experienced *higher* mean nominal wage growth than the above-median group (28.2% vs 21.5%; Welch t = −8.61, p < 0.001, Cohen's d = −0.67), and the non-parametric test agreed. Analysis 3, however, showed that this raw gap is largely not an AI story. High-exposure occupations are concentrated at the top of the wage distribution (r = 0.485 between AIOE score and 2023 median wage), and 2019–2025 was a period in which low-wage occupations received unusually large nominal raises due to pandemic-era labor shortages and minimum-wage increases. Once 2019 baseline wage is controlled for, the AIOE coefficient shrinks to a small negative association (≈ −0.9 percentage points per unit of AIOE, t = −2.33, p ≈ 0.02), while the baseline-wage control is a far stronger predictor in the same model (t = −11.61). Our first hypothesis therefore fails mostly because of a low-wage base effect, with at most a small residual negative association between exposure and wage growth.

For employment, our second hypothesis was clearly rejected. High-exposure occupations *grew*: mean employment growth was +4.2% versus −6.4% in the low-exposure group (p < 0.001, d = 0.42), and in the continuous model one unit of AIOE was associated with roughly +5 percentage points of employment growth even after controlling for 2019 employment size. Importantly, Analysis 4 showed this advantage existed both before (2019–2022: +6.5 pp) and after (2022–2025: +3.8 pp) the release of ChatGPT, suggesting it reflects longer-run demand for analytical and information-processing work rather than a generative-AI-specific effect.

### Context within prior work

Our findings give empirical content to a caution that Eloundou et al. raised themselves: technical feasibility does not guarantee labor-market outcomes. Their task-level analysis, like the ability-level AIOE of Felten et al. that we use, measures what AI *could* affect. Through May 2025, that theoretical exposure had not translated into measurable aggregate employment decline in exposed occupations; if anything, the pattern is more consistent with the augmentation interpretation built into the AIOE framework, in which exposure captures information- and language-intensive work whose demand continued to grow. Our wage results, in turn, echo a well-documented macro phenomenon — pandemic-era wage compression at the bottom of the wage distribution — more than they reveal anything specific about AI.

### Limitations

Several limitations qualify these results. First, all growth rates are nominal: the 2019–2025 window includes the highest U.S. inflation in four decades. Because every occupation faces the same price level, deflating by a common CPI would not change cross-group comparisons, but the magnitudes should not be read as real wage gains. Second, our balanced-panel design introduces survivorship: occupations that disappeared, fell below reporting thresholds, or were reclassified between 2019 and 2025 are excluded, which could understate displacement precisely where it was most severe. Relatedly, we excluded 21 ever-top-coded (mostly physician) occupations and approximated the medians of three harmonized computer occupations with employment-weighted averages. Third, national occupation-level aggregates hide within-occupation change: AI could be reshaping who is hired (for example, entry-level versus experienced workers), task content, or regional demand without moving an occupation's total employment or median wage. Fourth, the AIOE index is a theoretical, ability-based exposure measure constructed before the current generation of AI models, and it captures potential exposure rather than actual adoption. Fifth, the post-generative-AI window is short — at most three OEWS reference dates after ChatGPT's release — and adoption lags are plausibly longer; OEWS also excludes self-employed and gig workers. Finally, this is an observational design: COVID-19, inflation, interest rates, and tech-sector layoffs all coincide with our window, so none of our estimates should be interpreted causally.

### Ethical considerations

The most important ethical issue remains interpretation. Our results should not be read as "AI-exposed jobs are safe," any more than the exposure literature should be read as "AI-exposed jobs will disappear": both are aggregate associations over a short, turbulent period. We deliberately report that our hypotheses were not supported rather than emphasizing only the significant patterns, and we avoid framings that could be used to justify discriminatory hiring decisions or to steer individuals toward or away from careers on the basis of an exposure score alone. Because national aggregates conceal demographic and regional differences, our results also cannot speak to whether AI's effects are distributed equitably across workers within occupations.

### Future work

Natural next steps are to extend the panel as new OEWS releases arrive, to repeat the analysis with CPI-deflated real wages, and to complement the binary median split with continuous and quantile-based specifications. More fundamentally, theoretical exposure should be paired with measures of actual adoption — for example, job postings mentioning AI tools or firm-level adoption surveys — enabling difference-in-differences or event-study designs around late 2022. Within-occupation analyses using the OEWS percentile wages (10th–90th) and state-level data could reveal distributional and regional effects that national medians hide, and comparing alternative exposure indices, such as Eloundou et al.'s task-level scores, would test whether our conclusions are robust to the choice of exposure measure.

### Conclusion

Between 2019 and 2025, U.S. occupations with high theoretical exposure to language-model AI saw faster employment growth and slower nominal wage growth than less-exposed occupations — and the wage gap is mostly a low-wage base effect rather than an AI effect. Theoretical exposure has not yet translated into aggregate labor-market displacement. Whether that reflects augmentation, slow adoption, or simply how early in the AI era this window falls is a question our design cannot answer, and one that the next several years of data will.
